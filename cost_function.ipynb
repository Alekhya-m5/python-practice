```python
import numpy as np
import matplotlib.pyplot as plt
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
plt.scatter(x_train,y_train)
```

```python
def compute_cost(x, y, w, b): 
    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
total_cost=compute_cost(x_train, y_train, 200, 100)
print(total_cost)

w =200
b = 100

x_line = np.linspace(0,3,100)
y_line = w * x_line + b

plt.scatter(x_train, y_train)
plt.plot(x_line, y_line)
plt.show()
```

```python
w_values = np.linspace(-100,100,100)
cost_values = []

for w in w_values:
    cost = compute_cost(x_train, y_train, w, 0)
    cost_values.append(cost)

plt.plot(w_values, cost_values)
plt.xlabel("w")
plt.ylabel("Cost")
plt.show()
```
